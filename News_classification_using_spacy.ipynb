```python
import pandas as pd
df = pd.read_csv("https://github.com/codebasics/nlp-tutorials/raw/main/14_word_vectors_spacy_text_classification/Fake_Real_Data.csv")
df.head()
```

```python
df['label_num'] = df['label'].map({'Fake':0,'Real':1})
df.head()
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
df['vector'] = df['Text'].apply(lambda x: nlp(x).vector)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =train_test_split(
    df.vector.values,
    df.label_num,
    test_size = 0.2,
    random_state = 2022)
```

```python
import numpy as np
X_train_2D = np.stack(X_train)
X_test_2D = np.stack(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report

#1. Scaling object for removing the negative values
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_2D)
X_test_scaled = scaler.transform(X_test_2D)

#2. creating a KNN model object
clf = MultinomialNB()

#3. fit with all_train_embeddings and y_train
clf.fit(X_train_scaled,Y_train)

#4. get the predictions for all_test_embeddings and store it in y_pred
Y_pred = clf.predict(X_test_scaled)

#5. print the classfication report
print(classification_report(Y_test,Y_pred))
```

```python
from  sklearn.neighbors import KNeighborsClassifier

#1. creating a KNN model object
clf = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

#2. fit with all_train_embeddings and y_train
clf.fit(X_train_2D, Y_train)

#3. get the predictions for all_test_embeddings and store it in y_pred
Y_pred = clf.predict(X_test_2D)

#4. print the classfication report
print(classification_report(Y_test, Y_pred))
```
